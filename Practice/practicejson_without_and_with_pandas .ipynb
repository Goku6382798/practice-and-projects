```python
import json
import glob
import pandas as pd

def get_column_names(ds_name, file_path):
    with open(file_path, "r") as json_file:
        data = json.load(json_file)
    ku = data.get(ds_name, [])  # Avoids KeyError if ds_name is missing
    sorted_ku = sorted(ku, key=lambda col: col['column_position'])
    xyz2 = [k['column_name'] for k in sorted_ku] 
    return xyz2

def create_df(ds_name):
    schema_file = '/home/gautam/airflow/retail_db/schemas.json'
    column_names = get_column_names(ds_name, schema_file)
    
    file_pattern = f'/home/gautam/airflow/retail_db/{ds_name}/part-*'
    files = glob.glob(file_pattern)  # Get list of matching files
    
    if not files:
        raise FileNotFoundError(f"No files found for {file_pattern}")
    
    df = pd.concat([pd.read_csv(file, names=column_names, header=None) for file in files], ignore_index=True)
    
    return df
```

```python
# ✅ Merge orders_df and order_items_df
merged_df = orders_df.merge(order_items_df, left_on='order_id', right_on='order_item_order_id')

# ✅ Calculate total revenue per customer
customer_spending = merged_df.groupby('customer_id')['order_item_subtotal'].sum().reset_index()

# ✅ Count total orders per customer
customer_orders = merged_df.groupby('customer_id')['order_id'].nunique().reset_index()

# ✅ Merge both DataFrames to get spending and order count together
customer_metrics = customer_spending.merge(customer_orders, on='customer_id')

# ✅ Calculate Average Order Value (AOV)
customer_metrics['average_order_value'] = customer_metrics['order_item_subtotal'] / customer_metrics['order_id']

# ✅ Rename columns for clarity
customer_metrics.rename(columns={'order_item_subtotal': 'total_spending', 'order_id': 'total_orders'}, inplace=True)

customer_metrics[:10]
```

Fill with median	Best for skewed numerical data	        df.fillna(df.median())
Fill with mode	    Best for categorical columns	        df.fillna(df.mode()[0])
Forward fill	    Use previous row’s value	            df.fillna(method='ffill')
Backward fill	    Use next row’s value	                df.fillna(method='bfill')
